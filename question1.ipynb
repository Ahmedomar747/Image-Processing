```python
import matplotlib.pyplot as plt 
import numpy as np
from PIL import Image

img = np.asarray(Image.open("Sphinx.png"))

Width = img.shape[1]
Height = img.shape[0]

H = np.zeros(256,dtype=int)
Hc = np.zeros(256,dtype=int)
EH = np.zeros(256,dtype=int)

new_img = np.zeros(img.shape)

for _,w in enumerate(img):
    for _,v in enumerate(w):
        H[v] += 1

P = np.divide(H,(Width * Height))
Pc = np.cumsum(P)

EH = np.round(np.multiply(Pc,255))



#plt.savefig('Eq.jpg')

```

```python
for c,n in enumerate(EH):
    for i,row in enumerate(img):
        for j,val in enumerate(row):
            if img[i,j] == c:
                new_img[i,j] = n
        
        
plt.imshow(new_img,cmap="gray")
```

```python
histo=np.zeros(256)
for i in range(0,Width):
    for j in range(0,Height):
        colour=(img[j,i])
        histo[colour]=histo[colour]+1
        
plt.plot(histo)    
plt.savefig('Before_Eq.jpg')
```
